```python
import numpy as np
import pylab as py
%matplotlib inline
```

```python
D = 3
a = 2
Nt = np.arange(0.05, 50.0, 0.01)
N = Nt + a**2/(2*np.pi) + a * np.sqrt(8*Nt/np.pi - (D-2)/(3*np.pi))
N2 = Nt + a**2/(2*np.pi) - a * np.sqrt(8*Nt/np.pi - (D-2)/(3*np.pi))
```

```python
py.plot(Nt, N, Nt, N2, ls='--')
py.fill_between(Nt, N, N2, color="none", hatch="X", edgecolor="b", linewidth=0.0)
py.text(25, 25, "hello", bbox=dict(facecolor='white', ec='none'))

py.xlabel(r'$N_L$')
py.ylabel(r'$N_R$')
py.title(r'E = 0')
py.ylim((0, 50))
```
